```python
import pandas as pd
path="/content/Customer Purchasing Behaviors.csv"
df=pd.read_csv(path)
df
```

```python
x=(df["annual_income"])
y=(df["purchase_amount"])
print(x,y)
```

```python
mean_x=sum(x)/len(x)
mean_y=sum(y)/len(y)
print(mean_x,mean_y)
```

```python
numer=0
denom=0
for i in range(len(x)):
  numer+=(x[i]-mean_x)*(y[i]-mean_y)
  denom+=(x[i]-mean_x)**2
  m=numer/denom
  c=mean_y-(m*mean_x)
print(m,c)
```

```python
import matplotlib.pyplot as plt
import numpy as np
max_x=np.max(x)+5
min_x=np.min(x)-5
a=np.linspace(min_x,max_x)
b=m*a+c
plt.plot(a,b,color='red')
plt.scatter(x,y,color='blue')
plt.xlabel("annual_income")
plt.ylabel("purchase_amount")
plt.show()
```
